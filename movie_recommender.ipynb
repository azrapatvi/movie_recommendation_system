types of recommendation systems :

1.   Content Based : this is basically a type of recommendation system that recommends the items based on past data measn based on your past behaviour it recommends the object
2.   Collaborative Based : this is another type of recommender system that recommends you items based on other person who is similar to you measn lets say a and b both gave 5 rating to a movie then they are considierd to be similar to each other and next time when any one watches another movie then that gets recommende to the other person
1.   hybrid :this is the combination of both recommender systems content based and collaborative based and netflix uses this
here basically system checks your past behaviour also people similar to you and Then gives a final smart recommendation

```python
# literal_eval is a fucntion that can convert the string of list into list not only it will convert to list but can convert to any data type(like list,tuple,dict etc)

# ex:if you str1='[1,2,3]'
# and if you do list(str1) then output will be like this ['[', '1', ',', ' ', '2', ',', ' ', '3', ']']
# but literal_eval will convert it into proper list that is [1,2,3]

from ast import literal_eval

str1='[1,2,3]'
print(type(str1))
print(type(literal_eval(str1)))
print(literal_eval(str1))
```

```python
# helper function to get name of genres
import ast

def convert(obj):
  l1=[]
  for i in ast.literal_eval(obj): # loops through all the dictionary present in list
    l1.append(i['name'])

  return l1
```

```python
convert(movies.iloc[0]['genres'])
```

```python
# as now i have creaed a function and wants to apply it on all the rows of df so apply()

movies['genres']=movies['genres'].apply(convert)

movies.head(2)
```

```python
def fetch_director(obj):
  l2=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l2.append(i['name'])
      break
  return l2
```

Bag of Words

```python
from nltk import PorterStemmer

stemmer=PorterStemmer()
```

```python
stemmer
```

```python
def stemming(text):
  y=[]

  for i in text.split():
    y.append(stemmer.stem(i))

  return " ".join(y)

```

```python
df['tags']=df['tags'].apply(stemming)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')

cv
```

```python
cv.fit_transform(df['tags']).toarray()
```

```python
vectors=cv.fit_transform(df['tags']).toarray() # each movie converted into a vector
vectors
```

```python
vectors.shape
```

```python
features=cv.get_feature_names_out()
features[:20]
```

```python
len(cv.get_feature_names_out())
```

```python
# cosine similarity measn cosine distance between two movies this is not like euclidiean distance here the angle is a distnace between towo movies
```

```python
# cosine distance of each movie with each other movie
#1 ->high similarity
#0 ->no similarity

from sklearn.metrics.pairwise import cosine_similarity

similarity=cosine_similarity(vectors) # this calculates distance of 1st movie with other 4806 movies then 2nd movie with other 4806 movie and so on
```

```python
def recommend(movie):
  movie_index=df[df['title']==movie].index[0]
  distances=similarity[movie_index]

  sorted_movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in sorted_movie_list:
    print(df.iloc[i[0]]['title'])
    #print(i[0]) # gives index of top 5 similar movies
    print()
```

```python
recommend('Avatar')
```
